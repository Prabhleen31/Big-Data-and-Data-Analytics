```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assume data is a DataFrame containing your data

# List of feature columns
feature_columns = ['Price(Rs)', 'DIMENSIONS_Length()', 'DIMENSIONS_Width(mm)', 'DIMENSIONS_Height(mm)','DIMENSIONS_Wheelbase(mm)']  # Add more feature names

# Create subplots
num_plots = len(feature_columns)
num_cols = 2  # Number of columns in the subplot grid
num_rows = (num_plots + num_cols - 1) // num_cols  # Calculate the number of rows needed

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 6))

# Flatten the axes array if there's only one row
if num_rows == 1:
    axes = [axes]

# Loop through each feature and plot its distribution
for i, col in enumerate(feature_columns):
    row_idx = i // num_cols
    col_idx = i % num_cols
    ax = axes[row_idx][col_idx]

    sns.distplot(data[col], ax=ax, color='blue')
    ax.set_title(f'Distribution of {col}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```
